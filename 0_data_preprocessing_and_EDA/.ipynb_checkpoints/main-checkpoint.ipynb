### Data Preprocessing

1. Data Cleaning
    - Missing Values
        - remove
        - fill in with mean, median, mode
        - fill in with interpolation value
    - Inconsistent Values / Outliers
        - Observe with EDA
    - Duplicate Values
    
    
2. Data Transformation (Feature Encoding)
    - Categorical
        - One-hot encoding
        - Label encoding
    - Numeric
        - Scaling (ex: Standardlization, MinMaximum)


3. Data Reduction (減少雜訊)
    - Dimensionality Reduction
        - Principal Component Analysis
        - Singular Value Decomposition


4. Feature Sampling
    - Stratified Sampling (when dataset size > memory)


4. Data Split
    - Train/Validation/Test

        

> Not all the steps are applicable for each problem, it is highly dependent on the data we are working with. <br>

```
    [Cosmetics-Shop-2019-Dec.csv] Events log from a Cosmetics Shop
    Domain Knowledge:
    1. EC buyers may have different degree of view/cart/purchase frequency
    2. An EC-DS scenario: 
        A female user signed up, after she browse for a while, our model real-time give her a *classification; Within the taxonomy, our model already learned from historical data:
        - This kind of user have the highest probability to "cart" after 8 times of "view"
        - This kind of user have the highest probability to "purchase" after 3 times of "cart"
        - We can count the *probability of the user "purchase" by her cumulated events log.
        - *The probability decreases when days of user leave-website increase.
        - If the user left website, here comes some marketing levers:
            - Active when leave website >=*1 days: reminder (email)
                - seen in 2day, back to website: 機率回到離站前水準
                - seen in 2day, didn't back to website: 機率繼續遞減，看到提醒後每一天多遞減 *10%
                - unseen in 2day: 用戶對消息很不靈敏，機率繼續遞減，看到提醒後每一天多遞減 *3%
                    - reminder (SMS) 利潤扣掉 SMS 成本
                        - seen in 1day, back to website: 機率回到離站前水準
                        - seen in 1day, didn't back to website: 機率繼續遞減，看到提醒後每一天多遞減 *20%
                        - unseen in 1day: 用戶對消息很不靈敏，機率繼續遞減，看到提醒後每一天多遞減 *6%
            - If the probability below the *threshold, the EC will deliver time-limited coupon to correspondent user:
                - discount: 5% - 30% off, by first/2+ purchase value that kind of user's historical transactions, make it more closer to a average purchase amount.

    Issues we're curious about:
    1. How we segment users by their events sequence?
    2. Within each segment, 
        - Before a user "cart", how many "view" will happen? / Before a user "purchase", how many "view"/"cart"/"remove_from_cart" will happen?
        - Consumption amount at 1st purchase / other repeat purchases
        - How to measure probability of purchase by their events log?
        - What's the least probability to purchase?
        
    
```

````python
def events_count(tmp_groupby, col):
    try:
        views_before_purchase = tmp_groupby.query('event_type == "view"')[col].values[0]
    except:
        views_before_purchase = 0
    
    try:
        carts_before_purchase = tmp_groupby.query('event_type == "cart"')[col].values[0]
    except:
        carts_before_purchase = 0
        
    try:
        removes_before_purchase = tmp_groupby.query('event_type == "remove_from_cart"')[col].values[0]
    except:
        removes_before_purchase = 0
    
    return np.array([views_before_purchase, carts_before_purchase, removes_before_purchase])


collect = list()

for userid in tqdm(cleaned.user_id.unique()):
    
    # userid = 576802932
    _collect = list()
    tmp = cleaned.query('user_id == @userid').sort_values('event_time')
    first_enter_timing = tmp.event_time.values[0]
    last_enter_timing = tmp.event_time.values[-1]
    purchases = tmp.query('event_type == "purchase"').shape[0]
    purchased_values = tmp.query('event_type == "purchase"').price.sum()
    purchased_u_products = tmp.query('event_type == "purchase"').product_id.unique().shape[0]
    try:
        first_purchase_timing = tmp.query('event_type == "purchase"').event_time.values[0]  # 篩第一次purchase
    except: # 沒有 purchase 事件
        first_purchase_timing = 'None'
    tmp = tmp.query('event_time <= @first_purchase_timing')
    enter_uni_dayhours = tmp['day-hour'].unique().shape[0]

    # events
    tmp_groupby = tmp.groupby(['event_type']).size().reset_index().sort_values('event_type')
    events_before_purchase = events_count(tmp_groupby, col=0)
        
    # values
    tmp_groupby = tmp.groupby(['event_type'])['price'].sum().reset_index().sort_values('event_type')
    values_before_purchase = events_count(tmp_groupby, col='price')
    
    # uni_products
    tmp_groupby = tmp.groupby(['event_type'])['product_id'].nunique().reset_index().sort_values('event_type')
    uni_products_before_purchase = events_count(tmp_groupby, col='product_id')
    
    # uni_categories
    tmp_groupby = tmp.groupby(['event_type'])['category_id'].nunique().reset_index().sort_values('event_type')
    uni_categories_before_purchase = events_count(tmp_groupby, col='category_id')
    
    # uni_product_values
    tmp_groupby = tmp[['event_type','product_id','price']].drop_duplicates().groupby(['event_type'])['price'].sum().reset_index().sort_values('event_type')
    uni_product_values_before_purchase = events_count(tmp_groupby, col='price')
    
    # uni_categories_values
    tmp_groupby = tmp[['event_type','category_id','price']].drop_duplicates().groupby(['event_type'])['price'].sum().reset_index().sort_values('event_type')
    uni_categories_values_before_purchase = events_count(tmp_groupby, col='price')

    for item in [userid, events_before_purchase, values_before_purchase, uni_products_before_purchase, uni_categories_before_purchase, 
                uni_product_values_before_purchase, uni_categories_values_before_purchase, purchases, purchased_values, 
                purchased_u_products, first_enter_timing, last_enter_timing, first_purchase_timing, enter_uni_dayhours]:
        _collect = np.append(_collect, item)
    
    collect.append(_collect)

df_collect = pd.DataFrame(collect, columns=['userid','views_before_purchase','carts_before_purchase','removes_before_purchase',
                                            'viewed_values_before_purchase','carted_values_before_purchase','removeed_values_before_purchase',
                                            'viewed_uni_products_before_purchase','carted_uni_products_before_purchase','removed_uni_products_before_purchase',
                                            'viewed_uni_categories_before_purchase','carted_uni_categories_before_purchase','removed_uni_categories_before_purchase',
                                            'viewed_uni_product_values_before_purchase','carted_uni_product_values_before_purchase','removed_uni_product_values_before_purchase',
                                            'viewed_uni_category_values_before_purchase','carted_uni_category_values_before_purchase','removed_uni_category_values_before_purchase',
                                            'purchases','purchased_values','purchased_uni_products',
                                            'first_enter_timing', 'last_enter_timing', 'first_purchase_timing', 'enter_uni_dayhours'])

df_collect.to_csv('output/Aggregated-byUser-Cosmetics-Shop-2019-Dec.csv', index=False)
df_collect.shape
df_collect.head()
````

````python
def count_duration(df, start_col, end_col, new_col_name):
    df['s'] = df[start_col].apply(lambda x: dt.datetime.strptime(str(x).replace(' UTC',''), '%Y-%m-%d %H:%M:%S'))
    df['e'] = df[end_col].apply(lambda x: dt.datetime.strptime(str(x).replace(' UTC',''), '%Y-%m-%d %H:%M:%S'))
    df[new_col_name] = df['e'] - df['s']
    df[new_col_name] = df[new_col_name].apply(lambda x: x.total_seconds() / 3600)
    df = df.drop(['s','e'], axis=1)
    
    return df

df_collect = count_duration(df_collect, start_col='first_enter_timing', end_col='last_enter_timing', new_col_name='total_duration_hr')
df_collect_ = df_collect.query('first_purchase_timing != "None"')
df_collect_ = count_duration(df_collect_, start_col='first_enter_timing', end_col='first_purchase_timing', new_col_name='total_purchase_duration_hr')
df_collect = pd.concat([df_collect.query('first_purchase_timing == "None"'), df_collect_])
df_collect.head()
df_collect.corr()
````
